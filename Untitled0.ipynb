```python
# 예시 그래프
graph = {
    '영서' : ['정원', '윤채'],
    '정원' : ['채완'],
    '윤채' : [],
    '채완' : []
}

# DFS 방식
def dfs(graph, v, visited):
  visited.add(v)
  print(v, end=' ')
  for neighbor in graph[v]:
    if neighbor not in visited:
      dfs(graph, neighbor, visited)
print("DFS 탐색 순서 : ")
dfs(graph, '영서', set())

# BFS 방식
from collections import deque

def bfs(graph, start):
  visited = set()
  queue = deque([start])
  visited.add(start)

  while queue:
    v = queue.popleft()
    print(v, end=' ')
    for neighbor in graph[v]:
      if neighbor not in visited:
        visited.add(neighbor)
        queue.append(neighbor)

print("BFS 탐색 순서 : ")
bfs(graph, '영서')

# 친구 추천 시스템(BFS 응용)
from collections import  deque

def recommend_friend(graph, start):
  visited = set()
  queue = deque([(start, 0)])

  print(f"{start}의 친구 추천 (2단계) : ")

  while queue:
    current, depth = queue.popleft()
    if current not in visited:
      visited.add(current)
      if depth == 2:
        print("-", current)
      if depth < 2:
        for neighbor in graph[current]:
          queue.append((neighbor, depth + 1))
recommend_friend(graph, '영서')
```

```python
from collections import deque

# 인접 리스트 형태의 그래프
graph = {
    "Alice": ["Bob", "Charlie"],
    "Bob": ["Alice", "David", "Eve"],
    "Charlie": ["Alice", "Frank"],
    "David": ["Bob"],
    "Eve": ["Bob", "Grace"],
    "Frank": ["Charlie", "Heidi"],
    "Grace": ["Eve"],
    "Heidi": ["Frank", "Ivan"],
    "Ivan": ["Heidi", "Judy"],
    "Judy": ["Ivan"]
}

def bfs_infection(graph, start, max_depth=2):
    visited = {start}
    queue = deque([(start, 0)])  # (노드, 거리)
    infected = []

    while queue:
        node, depth = queue.popleft()

        # 2단계까지만 감염 확산
        if depth >= max_depth:
            continue

        for neighbor in graph.get(node, []):
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append((neighbor, depth + 1))
                infected.append((neighbor, depth + 1))  # 감염자와 거리 기록

    return infected


infected_list = bfs_infection(graph, "Alice", max_depth=2)

print("=== 2단계 이내 감염 경로 ===")
for person, dist in infected_list:
    print(f"{person} (거리 {dist})")
```
